```python
# ndarray shape 변경하기 - ravel()

# 다차원 배열을 1차원 배열로 변경

# 원본 데이터를 변형.  (차원 변형은 없지만, 데이터는 바뀔 수 있음.)

# 참조변수 temp도 x와 같은 데이터 공간을 가리키기 때문.

# np.arange(15)는 0부터 14까지의 배열.
x = np.arange(15).reshape(3,5)
print(x)
```

```python
 print(np.ravel(x))	# row(행) 우선
```

```python
print(np.ravel(x, order='F'))	# order를 False -> field 우선
```

```python
 print(x)	# 결과는 바뀌지 않음. -> ravel()은 원본의 차원에는 영향을 미치지 않음.
```

```python
temp = x.ravel()	# row 우선으로 1차원 배열
print(temp)
```

```python
temp[0] = 100	# 참조변수가 같은 데이터 공간을 가리키기 때문에 원본도 함께 변경
```

```python
print(temp)
```

```python
print(x) # 여전히 1차원 배열이지만, x[0]은 100으로 변함. (원본 데이터 변형)
```

```python
# flatten()

# 다차원배열을 1차원 배열로 변경
#- copy를 생성하여 변경함. (원본이 변하지 않음.)
```

```python
# N차원 정렬

# axis : 차수를 나타내는 기준

# 행 데이터를 가리킬 때는 axis=0  -> 행 정렬 (위에서 아래로)

# 열 데이터를 가리킬 때는 axis=1 -> 열 정렬 (왼쪽에서 오른쪽으로)

# 매트릭스 데이터를 가리킬 때는 axis=2

arr2d = np.array([[5,6,7,8],[4,3,2,1],[10,9,12,11]])
print(arr2d, arr2d.shape)
```

```python
print(np.sort(arr2d, axis=1))	# axis=1 ; 각 행 내에서 정렬(열 정렬)
```

```python
print(np.sort(arr2d, axis=0))	# axis=0 ; 각 행끼리 정렬(행 정렬)
```

```python
# 행렬의 계산

# 덧셈, 뺄셈, 곱셈

# 행렬끼리의 덧셈, 뺄셈, 곱셈 계산

a=np.array([[1,2,3],[4,5,6]])
b=np.array([[3,4,5], [1,2,3]])

# 덧셈
print(a+b)
```

```python
# 뺄셈
print(a-b)
```

```python
# 곱셈
print(a*b)
```

```python
# 스칼라 곱셈도 가능함.
print(2*a)
```

```python
# dot product

# 첫번째 배열의 열과 두번째 배열의 행이 같을 경우, 곱셈이 가능.

a = np.array([[1, 2, 3], [1, 2, 3], [2, 3, 4]])

b = np.array([[1, 2], [3, 4], [5,6]])
a.shape, b.shape
```

```python
# 첫번째 배열의 열과 두번째 배열의 행이 같은 길이기 때문에 곱셈 가능.

# np.dot 메서드로 곱셈 가능.

np.dot(a,b)
```

```python
a.dot(b)	# 위와 같은 결과를 얻을 수 있음
```

```python
# Broadcasting

# 두 배열이 동일한 길이인 경우

a = np.array([[1,2,3], [2,3,4]])
b = np.array([[3,3,3], [3,3,3]])
print(a+b) # # 같은 길이일 경우, 같은 위치의 원소들끼리 더해준다.
```

```python
# 다중 차트
# 한 화면에 차트 그리기
data1 = np.arange(1,10) # 1, 10 사이의 배열 생성
plt.plot(data1) # data1 차트 설정

data2 = np.arange(2,12) # 2, 12 사이의 배열 생성
plt.plot(data2) # data2 차트 설정

data3 = np.arange(3, 14) # 3, 14 사이의 배열 생성
plt.plot(data3) # data3 차트 설정

plt.show() # 화면에 차트 출력
```

```python
# 여러 화면에 차트 그리기
data1 = np.arange(100,201)
plt.plot(data1)

data2 = np.arange(200,301)
plt.figure() # 한 개의 figure 추가 생성
plt.plot(data2)

plt.show()
```

```python
# subplot() 메서드를 이용한 차트 그리기
data1 = np.arange(100,201)
plt.subplot(1, 2, 1) # 행 1개, 열 2개, data1 차트는 1번째 인덱스에 위치
plt.plot(data1)

data2 = np.arange(200,301)
plt.subplot(1, 2, 2) # 행 1개, 열 2개, data2 차트는 2번째 인덱스에 위치
plt.plot(data2)

plt.show()
```
